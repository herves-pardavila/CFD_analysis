```python
from Temporal_Analysis import *
import matplotlib.pyplot as plt
```

```python
path_to_folder=r"C:/TFM_DAVIDH_2025/Tablas/"
```

```python
clase=Temporal_Analysis(path=path_to_folder, ustar="ustar0.44_", wind_dir="NE_", field="PS_T_V_", part="threshold_plane_z=3m",
                            coords=["X (m)", "Y (m)"])
```

```python
variables_eje1=["PS2"]
variables_eje2=["Temperature (K)"]
```

```python
xaxis_name="Time"
yaxis_name=r"Concentration ()"
ytwin_axis_name="Temperature (k)"
title="Passive scalar concentration"
font_size=20 # tamaño de fuente para título, nombres de los ejes, etiquetas de los ejes, leyenda, etc
fig_size=(40,10) #ancho-alto de la gráfica
cap_size=15 #longitud horizontal de los sombreritos de las barras de error
cap_thick=10 #anchura de los sombreritos de las barras de error
line_width=3 #ancho de linea
marker_size=15 #tamaño del punto

ylim=(2,2.6)

#name for saving
#fig_name=path_to_folder+"/borrador_graficas/"+ustar+viento+part+"_spatial_mean_PS"
#formato=".png"
```

```python
fig=plt.figure(figsize = fig_size)
ax=fig.add_subplot(111)
axtwin=ax.twinx()
fig.subplots_adjust(bottom=0.15)

ax.set_ylabel(yaxis_name,fontsize=font_size)
ax.set_xlabel(xaxis_name,fontsize=font_size)
ax.tick_params(axis='both', which='both', labelsize=font_size)
#ax.set_ylim(ylim)
axtwin.set_ylabel(ytwin_axis_name,fontsize=font_size)
axtwin.tick_params(axis='both', which='both', labelsize=font_size)
#ax.set_ylim(ylim)


for var in variables_eje1:
    clase.load_data(var)
    ax.errorbar(x=clase.hours,y=clase.means,yerr=clase.stds,fmt="-o",label=var,capsize=cap_size,lw=line_width,markersize=marker_size)


for var in variables_eje2:
    clase.load_data(var)
    axtwin.errorbar(x=clase.hours,y=clase.means,yerr=clase.stds,fmt="-s",label=var,capsize=cap_size,lw=line_width,markersize=marker_size)

fig.legend(fontsize=font_size)
```
